```python
import numpy as np
from Sarsa import SarsaAgent
from Qlearning import QlearningAgent
from Cliffworld import CliffworldEnv
import matplotlib.pyplot as plt
```

```python
size = 14
```

```python
max_episodes = 500
episode = 0
R = 0
steps = 0
fallSarsa = 0
```

```python
stepsPerEpisode = np.empty(max_episodes)
episodeTrajectoriesSarsa = []
trajectory = []
rewardsSarsa = []
```

```python
env = CliffworldEnv(size)
agent = SarsaAgent(stateSpace=(size // 3,size), epsilon=0.1)
state = env.reset()

while episode < max_episodes:
    trajectory.append(state)
    action = agent.move(state)
    state2, reward, terminated = env.step(action)
    R += reward
    action2 = agent.move(state2)
    agent.update(state, action, reward, state2, action2)
    if not terminated:
        R += reward
    state =  state2
    action = action2
    trajectory.append(state)

    steps += 1
    if terminated:
        stepsPerEpisode[episode] =  steps
        episodeTrajectoriesSarsa.append(trajectory)
        rewardsSarsa.append(R)
        episode += 1
        if (episode % 15) == 0: print(f"episode {episode} done, steps {steps} were needed. Terminated at {state}")
        steps = 0
        R = 0
        trajectory = []
        state = env.reset()
        
env.close()
```

```python
def plot_trajectory(trajectories, episode):
    grid = np.zeros((size // 3, size))


    for pos in trajectories[episode]:
        grid[pos] = 1

    fig, ax = plt.subplots(figsize=(20, 5))


    for i in range(len(trajectories[episode]) - 1):
        curr_pos = trajectories[episode][i]
        next_pos = trajectories[episode][i + 1]
        delta = (next_pos[1] - curr_pos[1], next_pos[0] - curr_pos[0])
        ax.arrow(curr_pos[1], curr_pos[0], delta[0], delta[1],
             head_width=0.05, head_length=0.05, fc='k', ec='k', linewidth=3)
        
    ax.set_xticks(np.arange(size))
    ax.set_yticks(np.arange(size//3))

    ax.set_title("Trajectory")
    ax.grid(True)
    plt.show()

```

```python
plot_trajectory(episodeTrajectoriesSarsa, np.argmax(rewardsSarsa))
```

```python
plot_trajectory(episodeTrajectoriesQL, np.argmax(rewardsQL))
```

```python
plot_trajectory(episodeTrajectoriesES, np.argmax(rewardsES))
```

### Remarks

What are $\epsilon$-soft on-policy methods really learning? Are they simply learning how to chose the optimal policy given the fact that it is $\epsilon$-soft? Meaning, the optimal policy given the fact that there will be random actions? More specifically, it sets the behaviour policy equal to the target policy. 

Q-learning on the other hand is learning the *real* optimal policy. It sets the target policy equal to the optimal policy.

Quote from Sutton and Barto:

*The on-policy approach in the preceding section is actually a compromise—it learns action values not for the optimal policy, but for a near-optimal policy that still explores.*

Expected SARSA subsumes both Q-learning and SARSA. In the implementation here the update and the action selection are both doing with the same probabilities so the policies are identical. If you change the weighting of these vector you get an off-policy method. In the limit for Q-learning this vector is 0 everywhere except on the action with the highest Q-value.

Finally double-Q learning helps with overestimation:

Q-values in Q-learning are driven by the maximum (greedy). In noise cases this can lead to bias. We can reduce this bias by using two sets for updating. The argmax is chosen from the Q-value you want to update but the 'look-up' is done in a second one. This makes you more robust to overestimation. Intuitively, if there's a noisy example in 1 of the 2 it'll get cancelled out by looking at the second Q-table.
